## 2. Clean data 

* multi-level headers
* messy Excel structure
* renamed columns
* forward-filled regions
* removed “-” values
* converted to numeric types
* built `df_clean`

```python
# reload while skipping title row. tells pandas: use TWO header rows skip the giant title row. Results in cleaner multi-level column names.

df = pd.read_excel(
    "../data/rental_data.xlsx",
    sheet_name="Table 12",
    header=[0,1],
    skiprows=1
)

df.head()
```

```python
# forward fill regions. NaNs mean "same region as above"

df.iloc[:,0] = df.iloc[:,0].ffill()
```

```python
# rename columns

df = df.rename(columns={
    df.columns[0]: "Region",
    df.columns[1]: "Suburb"
})
```

```python
df.head()
```

```python
# cleaner column names. ('1 Bed Flat', 'Median') to 1 Bed Flat_Median

df.columns = [
    f"{col1}_{col2}" if "Unnamed" not in str(col1) else col2
    for col1, col2 in df.columns
]
```

```python
df.head()
```

```python
# rename columns 

df = df.rename(columns={
    "Unnamed: 0_level_1": "Region",
    "Unnamed: 1_level_1": "Suburb"
})
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
# For 4 Bed House_median, have to converts valid numbers → numeric values. And invalid values like "-" -> "NaN"

df["4 Bed House_Median"] = pd.to_numeric(
    df["4 Bed House_Median"],
    errors="coerce"
)
```

```python
df[["Suburb", "4 Bed House_Median"]].sort_values(
    by="4 Bed House_Median",
    ascending=False
).head(10)
```

```python
# Also the case with 2 Bed Flat_Median. Contains values like "-"

df["2 Bed Flat_Median"] = pd.to_numeric(
    df["2 Bed Flat_Median"],
    errors="coerce"
)
```

```python
df.groupby("Region")["2 Bed Flat_Median"].mean()
```

```python
# Replace "-" with NaN

df = df.replace("-", None)
```

```python
# Convert all numeric columns

for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
```

```python
# Calculate total number of missing values

df.isna().sum().sort_values(ascending=False)
```

```python
# Create df_clean with only Median columns

keep_cols = [
    "Region",
    "Suburb",
    "1 Bed Flat_Median",
    "2 Bed Flat_Median",
    "3 Bed House_Median",
    "4 Bed House_Median"
]

df_clean = df[keep_cols]
```

```python
# Long format (melt)

df_melted = df_clean.melt(
    id_vars=["Region", "Suburb"],
    var_name="Property Type",
    value_name="Median Rent"
)
```

```python
df_clean.sort_values("4 Bed House_Median").head(10)
```

```python
# My df_clean simply does NOT include "2 Bed House_Median". I actually do have "2 Bed House_Median" in the original dataset, but it was simply not included in df_clean earlier. That was issue.

df_clean.columns.tolist()
```

```python
# strip whitespace

df_clean.columns = df_clean.columns.str.strip()
```

```python
df_clean.columns.tolist()
```

```python
# Fix: rebuild df_clean properly. Then make sure everything is numeric.

df_clean = df[[
    "Region",
    "Suburb",
    "1 Bed Flat_Median",
    "2 Bed Flat_Median",
    "2 Bed House_Median",
    "3 Bed House_Median",
    "4 Bed House_Median"
]]
```

```python
# Make sure everything is numeric:

for col in df_clean.columns[2:]:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
```

```python
df_clean.groupby("Region")["2 Bed House_Median"].mean()
```

```python
df_clean.groupby("Region")[[
    "1 Bed Flat_Median",
    "2 Bed Flat_Median",
    "3 Bed House_Median",
    "4 Bed House_Median"
]].mean()
```

```python
# Removing records with these regions since they're not in Melbourne
regions_to_remove = ['Geelong', 'Ballarat', 'Bendigo', 'Other Regional Centres']

df_clean = df_clean[~df_clean['Region'].isin(regions_to_remove)]
```
